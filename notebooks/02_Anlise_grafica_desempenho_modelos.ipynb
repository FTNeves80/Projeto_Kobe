### Notebook oficial para analise grafica do desempenho dos modelos

É necessatio utilizar o kenel kedro proprio do projeto KOBE

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay , roc_curve, auc ,precision_recall_curve, average_precision_score
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

```

```python
# Exemplo: y_true e y_pred
y_true = [1, 0, 1, 1, 0, 1]
y_pred = [1, 0, 0, 1, 0, 1]

# Gerando a matriz de confusão
cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)

plt.title("Matriz de Confusão")
plt.xlabel("Classe Predita")
plt.ylabel("Classe Real")
plt.grid(False)
plt.show()
```

## Curva ROC e AUC

```python
# y_true: valores reais (0 ou 1)
# y_scores: probabilidades previstas para a classe positiva (ex: modelo.predict_proba(X)[:,1])

# Exemplo:
y_true = [0, 0, 1, 1]
y_scores = [0.1, 0.4, 0.35, 0.8]

fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8,6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--')  # linha de referência (modelo aleatório)
plt.xlabel('Taxa de Falsos Positivos (FPR)')
plt.ylabel('Taxa de Verdadeiros Positivos (TPR)')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

### Curva de Precisão vs Revocação (Precision-Recall Curve)

```python
# y_true: valores reais (0 ou 1)
# y_scores: probabilidades previstas para a classe positiva (modelo.predict_proba(X)[:,1])

# Exemplo:
y_true = [0, 0, 1, 1]
y_scores = [0.1, 0.4, 0.35, 0.8]

precision, recall, thresholds = precision_recall_curve(y_true, y_scores)
avg_precision = average_precision_score(y_true, y_scores)

plt.figure(figsize=(8,6))
plt.plot(recall, precision, label=f'Precisão vs Revocação (AP = {avg_precision:.2f})')
plt.xlabel('Revocação')
plt.ylabel('Precisão')
plt.title('Curva de Precisão vs Revocação')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```
